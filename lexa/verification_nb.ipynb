```python
import matplotlib.pyplot as plt

def draw_plots(pos_data, anchor_data, neg_data, pos_anchor, anchor, neg_anchor):

    plt.rcParams["figure.figsize"] = (8, 8)

    f, axarr = plt.subplots(3, 2)
    # f.tight_layout(pad=3.0)

    plt.subplots_adjust(left=0.1,
                        bottom=0.1,
                        right=0.9,
                        top=0.9,
                        wspace=0.8,
                        hspace=0.4)

    axarr[0,0].imshow(pos_data[0])
    axarr[0,1].imshow(pos_data[1])
    axarr[1,0].imshow(anchor_data[0])
    axarr[1,1].imshow(anchor_data[1])
    axarr[2,0].imshow(neg_data[0])
    axarr[2,1].imshow(neg_data[1])

    axarr[0,0].title.set_text(pos_anchor)
    axarr[0,1].title.set_text(pos_anchor)
    axarr[1,0].title.set_text(anchor['label'])
    axarr[1,1].title.set_text(anchor['label'])
    axarr[2,0].title.set_text(neg_anchor)
    axarr[2,1].title.set_text(neg_anchor)
    
    plt.show()
```

```python
dvd_data = Ego4DVideoFolder(root= '/iris/u/nivsiyer/ego4d/videos2', manifest_csv='/iris/u/nivsiyer/ego4d/videos2/manifest.csv', 
                                    clip_size=10, step_size=1, is_val=False,
                                    transform_pre=None, transform_post=None,
                                    augmentation_mappings_json=None, augmentation_types_todo=None,
                                    is_test=False, robot_demo_transform=None)
```

```python
for i in range(10):
    pos_data, anchor_data, neg_data, anchor, pos_anchor, neg_anchor = dvd_data.get_ims_labels()
    draw_plots(pos_data, anchor_data, neg_data, pos_anchor, anchor, neg_anchor)
    print()
    print()
```
